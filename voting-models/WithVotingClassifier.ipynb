```python
import numpy as np
import pandas as pd
from sklearn.kernel_approximation import RBFSampler
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (precision_score, recall_score,f1_score, accuracy_score,mean_squared_error,mean_absolute_error)
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import Normalizer
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import (precision_score, recall_score,f1_score, accuracy_score,mean_squared_error,mean_absolute_error, roc_curve, classification_report,auc)

traindata = pd.read_csv('kddtrain.csv', header=None)
testdata = pd.read_csv('kddtest.csv', header=None)

X = traindata.iloc[:,1:42]
Y = traindata.iloc[:,0]
C = testdata.iloc[:,0]
T = testdata.iloc[:,1:42]

scaler = Normalizer().fit(X)
trainX = scaler.transform(X)

scaler = Normalizer().fit(T)
testT = scaler.transform(T)


traindata = np.array(trainX)
trainlabel = np.array(Y)

testdata = np.array(testT)
testlabel = np.array(C)



#traindata = X_train
#testdata = X_test
#trainlabel = y_train
#testlabel = y_test

```

```python
print("-----------------------------------------LR---------------------------------")
modelLR = LogisticRegression()
modelLR.fit(traindata, trainlabel)

# make predictions
expected = testlabel
np.savetxt('classical/expected.txt', expected, fmt='%01d')
predicted = modelLR.predict(testdata)
proba = modelLR.predict_proba(testdata)

np.savetxt('classical/predictedlabelLR.txt', predicted, fmt='%01d')
np.savetxt('classical/predictedprobaLR.txt', proba)

y_train1 = expected
y_pred = predicted
accuracy = accuracy_score(y_train1, y_pred)
recall = recall_score(y_train1, y_pred , average="binary")
precision = precision_score(y_train1, y_pred , average="binary")
f1 = f1_score(y_train1, y_pred, average="binary")

print("accuracy")
print("%.3f" %accuracy)
print("precision")
print("%.3f" %precision)
print("racall")
print("%.3f" %recall)
print("f1score")
print("%.3f" %f1)
```

```python
from sklearn.ensemble import VotingClassifier

# add models
modelLR = LogisticRegression()
modelNB = GaussianNB()
modelDT = DecisionTreeClassifier()
modelAB = AdaBoostClassifier(n_estimators=100)
modelRF = RandomForestClassifier(n_estimators=100)

# VotingClassifier 
voting_clf = VotingClassifier(estimators=[
    ('LR', modelLR),
    ('NB', modelNB),
    ('DT', modelDT),
    ('Adaboost', modelAB),
    ('RF', modelRF)
], voting='soft')  # soft voting is used, so the class will be selected by average of the possibility values

# train models
voting_clf.fit(traindata, trainlabel)

# make guesses
predicted = voting_clf.predict(testdata)
proba = voting_clf.predict_proba(testdata)

np.savetxt('classical/predictedlabelVoting.txt', predicted, fmt='%01d')
np.savetxt('classical/predictedprobaVoting.txt', proba)

# examine the performance of the predictions
expected = testlabel
y_train1 = expected
y_pred = predicted
accuracy = accuracy_score(y_train1, y_pred)
recall = recall_score(y_train1, y_pred , average="binary")
precision = precision_score(y_train1, y_pred , average="binary")
f1 = f1_score(y_train1, y_pred, average="binary")


# accuracy = accuracy_score(testlabel, predicted)
# recall = recall_score(testlabel, predicted, average="binary")
# precision = precision_score(testlabel, predicted, average="binary")
# f1 = f1_score(testlabel, predicted, average="binary")

print("Prediction Performance with Voting Classifier:")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
